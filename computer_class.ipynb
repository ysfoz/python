```python

class computer():
    ses = 10
    def __init__(self, pc_durumu,marka,ozellikler = {'CPU' : 'intel i7', 'ram':'16 GB','GPU':'radeon 4 GB'},mevcut_donanim = ['ATX kasa', 'sony monitor','hp printer'],externel_donanim = ['klavye', 'mause'],ses1 = 10):
        self.pc_durumu = pc_durumu
        self.marka = marka
        self.ozellikler = ozellikler
        self.mevcut_donanim = mevcut_donanim
        self.externel_donanim = externel_donanim
        self.ses1 = ses1
    
    def open(self):
        if self.pc_durumu == 'kapali':
            print('pc aciliyor...')
        self.pc_durumu = 'acik'
        
    def close(self):
        if self.pc_durumu == 'acik':
            print('pc kapatiliyor ...')
        self.pc_durumu = 'kapali'

    def __str__(self):
        for i in self.ozellikler:
            x  = self.ozellikler.items()
        return f'{self.marka} PC\' ye ait donanimsal ozellikler:\n{x}\nPC\'ye ait mevcut donanim:\n{self.mevcut_donanim}\nPc\'ye ait externel donanim:{self.externel_donanim}'

    def new_hardware(self,hardware):
        print('yeni donanim ekleniyor...')
        self.mevcut_donanim.append(hardware)
        return self.mevcut_donanim

    
    def externel_ekle(self,externel_hrd):
        print('yeni bir exterel cihaz algilandi')
        self.externel_donanim.append(externel_hrd)
        return self.externel_donanim
    @staticmethod # bu metod ile init icerisinde tanimlanmiyan attribute ler tanimlanarak, bir fonksiyon gibi kullanilmasi saglanir.
    def ekran_parlakligi(ekran_parlaklg = 3): # ekran parlakligi burada yani metod icersinde tanimlandigi icin her calistiginda tekrar 3 olarak basliyor.
        print('ekran parlakligini yukseltmek icin "+" azaltmak icin "-" ye, cikmak icin "q" basiniz')
        while True:
            deger = input('ekran parlakligi')
            if (deger == "-"):  
                if (ekran_parlaklg != 0):
                    ekran_parlaklg -= 1
                    print("parlaklik:",ekran_parlaklg)
            elif (deger == "+"):
                if (ekran_parlaklg != 23):
                    ekran_parlaklg += 1
                    print("parlaklik:",ekran_parlaklg)
            else:
                print("ekran parlakligi seviyesi:",ekran_parlaklg)
                break

        return ekran_parlaklg

    @classmethod
    def ses_ayar(cls,ayar = 4): # tipki ekran ayari gibi buda her calistirildiginda 4 olarak basliyor.
        cls.ses = ayar
        print('Sesi yukseltmek icin "+" azaltmak icin "-" ye, cikmak icin "q" basiniz')
        
        while True:
            fonk = input('Ses ayari')
            if (fonk == "-"):  
                if (ayar != 0):
                    ayar -= 1
                    print("ses:",ayar)
            elif (fonk == "+"):
                if (ayar != 23):
                    ayar += 1
                    print("Ses",ayar)
            else:
                print("Ses seviyesi:",ayar)
                break

        return ayar
    def ses_ayar2(self): # bu ise irek olarak init icersinde tanimlandigi icin son tanimlanan deger uzerinden devam eder.
        while True:
            fonk2 = input('Ses ayari 2')
            if (fonk2 == "-"):  
                if self.ses1 != 0:
                    self.ses1 -= 1
                    print("ses:",self.ses1)
            elif (fonk2 == "+"):
                if self.ses1 != 23:
                    self.ses1 += 1
                    print("Ses",self.ses1)
            else:
                print("Ses seviyesi:",self.ses1)
                break

        return self.ses1
    @classmethod # bu islem ile fonksiyon icerisinde tanimlanan bir degisken uzerinde degisiklik yapilir. bu degisiklik degiskenin kulllanildigi tum metodlari etkiler. bu fonksiyonda ses degiskeni metodlarda kullanilmadigi icin degisen bisey olmadi.
    def ses_ayar_3(cls,ayar):
        cls.ses = ayar
        

pc1 = computer(pc_durumu = 'acik', marka = 'apple')
pc2 = computer("kapali","hp")
print(10 * '*', 'Yapmak istedigiz islemi seciniz',10 * '*','\n\n"p" tusuna basarak Pc yi acabilir\n"s" tusuna basarak Pc yi kapatabilirsiniz\nPcye ait tum ozellikler icin "1"\nYeni donanim eklemek icin "2"ye \nYeni externel donanim eklemek icin "3"ye\nEkran parlakligi ayari icin "4" e \nSes ayarlari icin "5" e basiniz\nSes ayarlari2 icin "6" e basiniz\n\n'+ 53 *'*')
while True:
    islem = input('yapmak istediginiz islemei seciniz')
    if (islem == "q"):
        print("Programdan Çıkılıyor...")
        break
    elif islem == 'p':
        pc1.open()
    elif islem == 's':
        pc1.close()
    elif islem == "1":
        print(pc1)
    elif islem == "2":
        hardwr = input('eklemek isteginiz donanimi giriniz')
        pc1.new_hardware(hardwr)
    elif islem == "3":
        ext_hrdwr = input('eklemek isteginiz externel donanimi giriniz')
        pc1.externel_ekle(ext_hrdwr)
    elif islem == "4":
        pc1.ekran_parlakligi()
    elif islem == "5":
        pc1.ses_ayar()
    elif islem == "6":
        pc1.ses_ayar2()
    else:
        print('gecersiz islem')

x = (computer.ses_ayar_3(15))
print(computer.ses)
            
```
